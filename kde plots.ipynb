```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('length_and_incidence_by_project.csv', names = ['ProjectID', 'Length', 'Incidence'])
df['Counter'] = 1
df.head()
```

```python
g = sns.jointplot("Length", "Incidence", data=df, kind="kde", stat_func = None, xlim = (0, 40), ylim = (0,1))
g.fig
```

```python
h = sns.jointplot("Length", "Incidence", data=df, kind="hex", stat_func = None, xlim = (0, 40), ylim = (0,1))
h.fig
```
