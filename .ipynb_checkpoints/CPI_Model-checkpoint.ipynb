# Background:

An consulting organization seeks to predict the unit cost of a given product based on a number of variables in order to reduce the amonut of time they spend estimating supplier costs. Currently, when a request comes in, stakeholders contact a variety of vendors for cost estimates and combine this information with their own experience in order to provide an accurate cost estimate for their organization. 

# Variables

* DV
  * CPU: Cost Per Unit
* IV
  * Length: Product Length
  * Incidence: Availability of product
  * Region: Source of product
  * Language: Language of product
  
Note: All other variables are not reliable pieces of information

```r
## Source Data

source <- read.csv('cpi.csv')    
country_map <- read.csv('country_map.csv')
cols_to_drop = c('Methodology', 'Total.Quota', 'Max.Feasibility', 'Internet.Penetration', 'Recommended...Online', 'Translation.with.Edit', 'Comment.One', 'Comment.Two', 'Comment.Three', 'Country', 'OrigCountryCount')

head(source)
```

```r
features <- c('Length', 'Incidence')

intro_diagnostics <- function(x, y){
    print(y)
    print(paste('CORR', x, y, cor(source[[x]], source[[y]])))
    plot(source[[x]], source[[y]], main = y, xlab = x, ylab = y)
    hist(source[[y]], 20, main = y, xlab = y)    
}

bins = 20
for (f in features){
    intro_diagnostics('CPU', f)
}
intro_diagnostics(features[1], features[2])
```

## Key Observations:

* Length
  * Moderate correlation with CPU
  * Data seems to be rounded to nearest 5
  * Has relatively normal distribution
* Incidence
  * High inverse correlation with CPU
  * Slightly quadratic 
* No correlation between Length & Incidence

```r
## Data Transformations

drop_cols <- function(df, cols_to_drop){
    for (col in cols_to_drop){
        df[[col]] = NULL
    }
    return(df)
}

transform_df <- function(df){
    # Consolidate country into Region via Merge & simplify language
    df <- drop_cols(merge(source, country_map, on='Country'), cols_to_drop)
    df$Language <- ifelse(df$Language != 'English', 'Other', 'English')
    
    # Rescale incidence & make quadratic
    df$Incidence <- (df$Incidence) * 100
    df$I2 <- (df$Incidence ^ 2)
    
    # Capture Length/Incidence Interaction
    df$IncLength <- (df$Incidence * df$Length) / 10
    return (df)
}
```

```r
# Fit model

set.seed(998)
inTraining <- createDataPartition(source$CPU, p = .5, list = FALSE)
train_df <- transform_df(source[inTraining, ])
test_df <- source[-inTraining, ]

set.seed(42)

model <- train(
    CPU ~ ., train_df,
    method = 'lm',
    trControl = trainControl(
        method = 'cv', number = 20,
        repeats = 5, verboseIter = TRUE
    )
)
```

```r
print(model)
summary(model)
```

```r
plot(model$finalModel)
```

### We have a relatively good fitting model based on the CV. Let's take a look at the test df

```r
# Test model
#test1 <- test_df[1,]

predict_df <- transform_df(test_df)

p <- predict(model, predict_df, interval="prediction")
actuals <- predict_df$CPU
error <- (p - actuals)

sqrt(mean(error ^ 2))
```

```r
plot(p, actuals)
abline(lm(actuals ~ p))
plot(error, actuals)
```

### Still works pretty well! However, there are some key observations:

* When CPU is low, we have a very tight grouping of predicted vs actual values. As CPU goes up, we tend to have a higher residual. This makes sense given that the margin of error increases as we move away from the mean. Since our cost has a floor of >0, these high CPUs will always be outliers.

```r
confint(model$finalModel)
```

```r
bump <- p * 1.15
bump_error <- bump - actuals
plot(bump_error, actuals)
```
